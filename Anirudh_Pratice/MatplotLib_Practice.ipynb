```python
import numpy as np  

a = np.array([[10,2,3],[4,6,5],[8,7,9]])  

print(a)

print("Sorting along the rows:")  
print(np.sort(a))  

print("Sorting along the column:")  
print(np.sort(a, axis=0))  

data_type = np.dtype([('name', 'S10'),('marks',int)])  

arr = np.array([('Mukesh',200),('John',251)],dtype = data_type)  

print("Sorting data ordered by name")  

print(np.sort(arr,order = 'name'))  

print(np.sort(arr,kind='merge',order='name')) # By defualt kind is quick sort
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
'''If you select nrows=2 ,then we need to use two loops'''
fig,axes=plt.subplots(nrows=2,ncols=2)
for current_axis in axes:
    for ax in current_axis:
        ax.plot(x,y)
```

```python
'''You can also do like this.'''
cols=2
fig,axes=plt.subplots(nrows=1,ncols=cols)
for i in range(0,cols):
    axes[i].plot(x,y)
```

```python
'''If you have 3 rows then you don't need to use 3 loops'''
cols=3
rows=3
fig,axes=plt.subplots(nrows=rows,ncols=cols)
plt.tight_layout()
for current_axis in axes:
    for j in range (cols):
        current_axis[j].plot(x,y)
```

```python
'''If you have 3 rows then you don't need to use 3 loops'''
cols=2
rows=2
fig,axes=plt.subplots(nrows=rows,ncols=cols)
plt.tight_layout()
for i,current_axis in enumerate(axes):
    for j in range (cols):
        current_axis[j].set_title('Plot {} and Subplot {}'.format(i+1,j+1))
        current_axis[j].plot(x,y)
```

```python
# figsize=(width,height)
fig=plt.figure(figsize=(5,3),dpi=100) # dot per inch dpi
axes=fig.add_axes([0.5,1,1,1])
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Random')
axes.plot(x,y)
```

```python
'''If you have 3 rows then you don't need to use 3 loops'''
cols=2
rows=2
fig,axes=plt.subplots(nrows=rows,ncols=cols,figsize=(12,12))
plt.tight_layout()
for i,current_axis in enumerate(axes):
    for j in range (cols):
        current_axis[j].set_title('Plot {} and Subplot {}'.format(i+1,j+1))
        current_axis[j].plot(x,y)
```

```python
# Specify legends
# figsize=(width,height)
fig=plt.figure(figsize=(5,3),dpi=100) # dot per inch dpi
axes=fig.add_axes([0.5,1,1,1])
axes.plot(x,x**2,label='X vs X^2')
axes.plot(x,x**3,label='X vs X^3')
```

```python
# Specify legends
# figsize=(width,height)
fig=plt.figure(figsize=(5,3),dpi=100) # dot per inch dpi
axes=fig.add_axes([0.5,1,1,1])
axes.plot(x,x**2,label='X vs X^2',c='blue')
axes.plot(x,x**3,label='X vs X^3',c='orange')
axes.legend()
```

```python
# Change the location of legend
fig=plt.figure(figsize=(5,3),dpi=100) # dot per inch dpi
axes=fig.add_axes([0.5,1,1,1])
axes.plot(x,x**2,label='X vs X^2',c='blue')
axes.plot(x,x**3,label='X vs X^3',c='orange')
axes.legend(loc=6) # Best is 0
```
